### Numerical Optimizations

Consider the function $f(x, y) = (x^2-8x+7)/(3) + (y^2-5x+8)/(6)$.

1. Analytically (with pen and paper!), calculate the gradient **vector** in terms of x and y
2. Use matplotlib to plot this function as a 2-D **surface** and get a feel for where its minimum value is
2. Pick a random starting point (x, y) and calculate f
3. Pick a learning rate
4. Write code that starts at your starting point (x, y), then moves by your learning rate in the direction opposite the gradient (by plugging x and y into your calculated gradient)
5. Repeat until convergence and the minimum is found.

How do you know when the minimum is found aka when to stop?  Up to you. 


```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

%matplotlib inline
```

```python
X = np.arange(-20, 20,0.1)
Y = np.arange(-20, 20,0.1)
X, Y = np.meshgrid(X, Y)
F = (X**2 - 8*X + 7)/ 3.0 + (Y**2 - 5*X + 8) / 6.0
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, F,rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False);
ax.set_xlabel('X')
ax.set_ylabel('Y')


plt.show()
```

```python
def grad_F(x,y):
    return (2/3.0)*x-7/2.0, y/3.0
```

```python
def nextpoint(x,y,alpha):
    dX, dY = grad_F(x,y)
    x_next = x - alpha*dX
    y_next = y - alpha*dY
    return x_next, y_next
```

```python
# Select a random starting point, and use a learning rate of alpha = 0.1
xi,yi = 1, 1
alpha = 0.1


# Here I just made the Gradient descent stop after 1000 iterations
# You could make it stop when F(x,y) no changes very small each time
for i in xrange(1000):
    xi, yi = nextpoint(xi,yi,alpha)

print xi, yi
# you can see the optimal x, y values are (5.25, 0)
```
