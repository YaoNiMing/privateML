```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from dateutil import parser

%matplotlib inline
```

```python
os.chdir('/Users/Sam/ds/metis/metisgh/sf16_ds4/challenges/02-pandas')
df = pd.read_csv('2013_movies.csv')
df = df.dropna(how = 'any')
df = df.sort_values(['ReleaseDate'])
df
```

```python
dt_converter = lambda x: parser.parse(x)
df['ReleaseDate'] = df['ReleaseDate'].map(dt_converter)
```

```python
plt.figure(figsize=(20,3))
plt.scatter(np.array(df['ReleaseDate']),df['Budget'])
```

# Exercise 2.4

```python
pgdf = df[df['Rating'] == 'PG']
pg13df = df[df['Rating'] == 'PG-13']
rdf = df[df['Rating'] == 'R']
```

```python
plt.figure(figsize = (20,30))
plt.subplot(3,1,1)
plt.title('PG Movies', fontsize = 20)
plt.xlabel('Release Date', fontsize = 15)
plt.ylabel('Total Domestic Gross', fontsize = 15)
plt.scatter(np.array(pgdf['ReleaseDate']), pgdf['DomesticTotalGross'])

plt.subplot(3,1,2)
plt.title('PG-13 Movies', fontsize = 20)
plt.xlabel('Release Date', fontsize = 15)
plt.ylabel('Total Domestic Gross', fontsize = 15)
plt.scatter(np.array(pg13df['ReleaseDate']), pg13df['DomesticTotalGross'])

plt.subplot(3,1,3)
plt.title('R Movies' , fontsize = 20)
plt.xlabel('Release Date', fontsize = 15)
plt.ylabel('Total Domestic Gross', fontsize = 15)
plt.scatter(np.array(rdf['ReleaseDate']), rdf['DomesticTotalGross'])
```
