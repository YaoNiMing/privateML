```python
# 1.8
# Obtaining data for three other weeks and doing the same munging
def munge(df):
    df.columns = df.columns.str.strip()
    df = df.groupby(('C/A', 'UNIT', 'SCP', 'STATION', 'DATE', 'TIME')).sum().reset_index()
    df['ENTRIES_c'] = df.ENTRIES.diff().fillna(0)
    df.ENTRIES_c[(df['ENTRIES_c'] < 0) | (df['ENTRIES_c'] > 3900)] = 0
    df['DATETIME'] = df.DATE[df.index] + ' ' + df.TIME[df.index]
    df.DATETIME = df.DATETIME.apply(dateutil.parser.parse)
    df.DATE = df.DATE.apply(dateutil.parser.parse)
    df.TIME = df.TIME.apply(dateutil.parser.parse)
    df.TIME = df.TIME.apply(lambda x: x.time())
    df['Weekday'] = df.DATE.apply(datetime.datetime.weekday)
    return df

df1 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_160910.txt')
df2 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_160903.txt')
df3 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_160924.txt')

df1, df2, df3 = map(munge, (df1, df2, df3))
```

```python
dates_station = df[df.STATION == '1 AV'].groupby('Weekday').sum().reset_index().Weekday
counts_station = df[df.STATION == '1 AV'].groupby('Weekday').sum().reset_index().ENTRIES_c
dates_station1 = df1[df1.STATION == '1 AV'].groupby('Weekday').sum().reset_index().Weekday
counts_station1 = df1[df1.STATION == '1 AV'].groupby('Weekday').sum().reset_index().ENTRIES_c
dates_station2 = df2[df2.STATION == '1 AV'].groupby('Weekday').sum().reset_index().Weekday
counts_station2 = df2[df2.STATION == '1 AV'].groupby('Weekday').sum().reset_index().ENTRIES_c
dates_station3 = df3[df3.STATION == '1 AV'].groupby('Weekday').sum().reset_index().Weekday
counts_station3 = df3[df3.STATION == '1 AV'].groupby('Weekday').sum().reset_index().ENTRIES_c

plt.figure(figsize=(10,3))
plt.xticks(xrange(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.plot(dates_station2, counts_station2, label = 'Week of 9-03')
plt.plot(dates_station1, counts_station1, label = 'Week of 9-10')
plt.plot(dates_station, counts_station, label = 'Week of 9-17')
plt.plot(dates_station3, counts_station3, label = 'Week of 9-24')
plt.ylabel('Entries')
plt.xlabel('Weekday')
plt.legend()
```
