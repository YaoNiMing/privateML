```python
import pandas as pd
import numpy as np
import pydot
from IPython.display import Image  
import matplotlib.pyplot as plt
import pydotplus 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
```

```python
df = pd.read_csv('house-votes-84.data', header=None)
```

```python
df
```

```python
X = df.loc[:,1:]
X = X.replace('n',0)
X = X.replace('y',1)
X = X.replace('?', np.NaN)

ar1 = np.array(X)
replist = np.nanmean(ar1, axis=0)

for i in range(X.shape[1]):
    x = X.iloc[:,i]
    X.iloc[:,i].replace(np.NaN, replist[i], inplace=True)
    
```

```python
y = df.loc[:, 0]
y = y.apply(lambda x: 0 if x=='democrat' else 1)
dt = DecisionTreeClassifier()
dectree = dt.fit(X, y)
```

```python
dot_data = export_graphviz(dectree, out_file=None) 
graph = pydotplus.graph_from_dot_data(dot_data)
graph.write_pdf("dt.pdf") 
```

```python
# make categorical variables
def categorical(dataframe, categorical_column):
    uniq_categories = set(dataframe[categorical_column].tolist())
    for category in uniq_categories:
        dataframe[category] = 0
        dataframe[category] = dataframe[categorical_column].apply(lambda x:1 if x == category else 0)
    dataframe.drop(categorical_column, axis = 1, inplace=True)
    return True
```
