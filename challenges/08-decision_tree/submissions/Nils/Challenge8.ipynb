## 8.1 Decision Tree Classifier

```python
df = pd.read_csv('house-votes-84.data.txt', names = ['party', 'x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11',
                                                    'x12','x13','x14','x15','x16'])
```

```python
# clean up data

df.replace('n',0, inplace=True)
df.replace('y',1, inplace=True)
```

```python
means=[]
for i in xrange(16):
    means.append((df[df['x%s' %(i+1)]!='?'].iloc[:,i+1].mean()))
    
# replace ? by 0/1 with p = mean
def replace_no(x):
    if x == '?':
        x = np.random.binomial(1,means[i])
    return x

for i in xrange(16):
    df.iloc[:,i+1]=df.iloc[:,i+1].map(replace_no)
```

```python
X=df.iloc[:,1:]
y=df.iloc[:,0]
tr = tree.DecisionTreeClassifier()
```

```python
# Fit model


tr.fit(X,y)
```

```python
# plot tree

dot_data = tree.export_graphviz(tr, out_file=None) 
graph = pydotplus.graph_from_dot_data(dot_data) 
graph.write_pdf("party.pdf") 
```

## 8.2 Titanic kaggle decision tree

```python
df_train = pd.read_csv('train.csv')
```

```python
# clean data
df_train.drop(['Cabin','Ticket','Name','Embarked','PassengerId'], axis=1, inplace=True) # not usefull/ too many Nans
df_train.dropna(how='any', inplace=True)
```

```python
df_train.replace('male',1, inplace=True)
df_train.replace('female',0, inplace=True)
```

```python
X = df_train.drop(["Survived"], axis=1)
y = df_train["Survived"]
```

```python
tit_tr=tree.DecisionTreeClassifier()
tit_tr.fit(X,y)
```

```python
dot_data2 = tree.export_graphviz(tit_tr, out_file=None) 
graph = pydotplus.graph_from_dot_data(dot_data2) 
graph.write_pdf("survival.pdf") 
```
