```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('house-votes-84.data')

df

df = df.replace('y', 1)

df = df.replace('n', 0)
```

```python
# ex5.1

df1 = df.iloc[:,1:]
```

```python
df1
```

```python
# remove ?
sr1 = df1.replace('?', np.NaN).mean()
```

```python
for col in df1.columns:
    temp = df1.loc[:, col].replace('?', sr1[col])
    df1.loc[:, col] = df1.loc[:, col].replace('?', sr1[col])
```

```python
df1
```

```python
# ex5.2
from sklearn.model_selection import train_test_split

X = df1
y = df['republican']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 4444)
```

```python
# ex5.3
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn_accu = []
for k in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    knn_accu.append((k, accuracy_score(y_test, knn.predict(X_test))))
```

```python
knn_accu
```

```python
# ex5.4
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg_accuracy = accuracy_score(y_test, logreg.predict(X_test))
logreg_accuracy
```

```python
df['republican'].value_counts()
```

```python
# ex5.5
import matplotlib.pyplot as plt

%matplotlib inline
df['republican'].value_counts().plot.bar()
```

```python
# ex5.6
knn_accuracy = [accu[1] for accu in knn_accu]
knn_kvalue = range(1, 21)
fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.plot(knn_kvalue, knn_accuracy)

logreg_accuracy = [logreg_accuracy] * len(knn_kvalue)
ax.plot(knn_kvalue, logreg_accuracy, 'r')
```

```python
# ex5.7
from sklearn.model_selection import learning_curve
[m, train_err, ts_err] = learning_curve(logreg, X, y)

train_cv_err = np.mean(train_err, axis=1)
test_cv_err = np.mean(ts_err, axis=1)

fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.plot(m, train_cv_err, 'b')
ax.plot(m, test_cv_err, 'r')
```

```python
# ex5.7 knn
from sklearn.model_selection import learning_curve
knn = KNeighborsClassifier(n_neighbors=12)
[m, train_err, ts_err] = learning_curve(knn, X, y)

train_cv_err = np.mean(train_err, axis=1)
test_cv_err = np.mean(ts_err, axis=1)

fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.plot(m, train_cv_err, 'b')
ax.plot(m, test_cv_err, 'r')
```

```python
# ex5.8 

from sklearn.naive_bayes import GaussianNB
gau = GaussianNB()
gau.fit(X_train, y_train)
gau_accuracy = accuracy_score(y_test, gau.predict(X_test))
gau_accuracy
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
svc_accuracy = accuracy_score(y_test, svc.predict(X_test))
svc_accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt_accuracy = accuracy_score(y_test, dt.predict(X_test))
dt_accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf_accuracy = accuracy_score(y_test, rf.predict(X_test))
rf_accuracy
```

```python
# ex5.9
from sklearn.cross_validation import cross_val_score
print('knn ' + str(cross_val_score(knn, X, y).mean()))
print('logreg ' + str(cross_val_score(logreg, X, y).mean()))
print('gau ' + str(cross_val_score(gau, X, y).mean()))
print('svc ' + str(cross_val_score(svc, X, y).mean()))
print('dt ' + str(cross_val_score(dt, X, y).mean()))
print('rf ' + str(cross_val_score(rf, X, y).mean()))
```

```python
# ex5.10
df = pd.read_csv('house-votes-84.data')

df

df = df.replace('y', 1)

df = df.replace('n', 0)
```

```python
df.head()
```

```python
df1 = df.iloc[:, 1:]
```

```python
# replace ? with mode for df1
from collections import Counter

for col in df1.columns:
    ctr = Counter(df1[col])
    if ctr[0] > ctr[1]:
        df1.loc[:, col] = df1.loc[:, col].replace('?', 0)
    else:
        df1.loc[:, col] = df1.loc[:, col].replace('?', 1)
        
```

```python
X = df1.iloc[:, 1:]
y = df1.iloc[:, 0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 4444)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg_accuracy = accuracy_score(y_test, logreg.predict(X_test))
logreg_accuracy
```

```python
from sklearn.cross_validation import cross_val_score
print('logreg ' + str(cross_val_score(logreg, X, y).mean()))
```

```python
# ex5.11
dfm = pd.read_csv('2013_movies.csv')
```

```python
dfm.head()
```

```python
dfm['Rating'].value_counts().plot.bar()
```

```python
dfm = dfm.dropna()
X = dfm.iloc[:, [1,2,5]]
y = dfm.iloc[:, 4]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 4444)
```

```python
for k in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    print(k, accuracy_score(y_test, knn.predict(X_test)))
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg_accuracy = accuracy_score(y_test, logreg.predict(X_test))
logreg_accuracy
```

```python
logreg.coef_
```

```python
# baseline predictor
def baseline(X_train, y_train, X_test):
    ctr = Counter(y_train)
    pred = ctr.most_common()[0][0]
    y_predict = pd.Series([pred] * len(X_test), index = X_test.index)
    return y_predict
```

```python
y_predict = baseline(X_train, y_train, X_test)
baseline_accuracy = accuracy_score(y_test, y_predict)
baseline_accuracy
```
