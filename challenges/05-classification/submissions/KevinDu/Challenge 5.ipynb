# 5.1


```python
# 5.1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('/Users/kevin/Downloads/house-votes-84.data', names=['Party']+['X'+str(i) for i in xrange(1,17)])
df
```

```python
df.replace('y', 1, inplace=True)
df.replace('n', 0, inplace=True)
df.replace('?', np.nan, inplace=True)
df.fillna(df.mean(),inplace=True)
```

# 5.2

```python
# 5.2

from sklearn.cross_validation import train_test_split

X = df.drop(['Party'], axis=1)
y = df.Party
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4444)
```

# 5.3

```python
# 5.3

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

accuracy = []

for i in xrange(1,21):
    knn = KNeighborsClassifier(i)
    knn.fit(X_train, y_train)
    y_predict = knn.predict(X_test)
    accuracy.append(accuracy_score(y_test, y_predict))
```

```python
plt.plot(xrange(1,21), accuracy)
plt.xlabel('# of neighbors')
plt.ylabel('Test set accuracy')
plt.title('KNN Accuracy vs. # of Neighbors');
```

# 5.4

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, y_train)
y_predict = lr.predict(X_test)
lr_score = accuracy_score(y_test, y_predict)
lr_score
```

# 5.5

```python
# 5.5

sns.barplot(df.Party.value_counts().index, df.Party.value_counts());
```

```python
# Predict all Democrat
y_predict = len(X_test)*['democrat']
democrat_score = accuracy_score(y_test, y_predict)
democrat_score
```

```python
# Predict all Republican
y_predict = len(X_test)*['republican']
republican_score = accuracy_score(y_test, y_predict)
republican_score
```

# 5.6


```python
# 5.6

plt.plot(xrange(1,21), accuracy, label = 'knn')
plt.plot(xrange(1,21), [lr_score]*20, label = 'Logistic')
plt.plot(xrange(1,21), [democrat_score]*20, label = 'Democrat')
plt.plot(xrange(1,21), [republican_score]*20, label = 'Republican')
plt.legend(loc=0)
plt.xlabel('# of neighbors')
plt.ylabel('Test set accuracy')
plt.title('Model Accuracy vs. # of Neighbors');
```

# 5.7

```python
# 5.7

from sklearn.learning_curve import learning_curve

m, train_err, test_err = learning_curve(lr, X, y)

train_cv_err = np.mean(train_err, axis=1)
test_cv_err = np.mean(test_err, axis=1)
```

```python
train_score
```

```python
plt.plot(m, train_cv_err, label='Training Error')
plt.plot(m, test_cv_err, label = 'Test Set Error')
plt.title('Learning Curve of Logistic Regression')
plt.xlabel('Size of Training Observations')
plt.ylabel('Accuracy')
plt.legend(loc=0);
```

```python
m, train_err, ts_err = learning_curve(KNeighborsClassifier(np.argmax(accuracy)+1), X, y)

train_cv_err = np.mean(train_err, axis=1)
test_cv_err = np.mean(ts_err, axis=1)
```

```python
plt.plot(m, train_cv_err, label='Training Error')
plt.plot(m, test_cv_err, label = 'Test Set Error')
plt.title('Learning Curve of Logistic Regression')
plt.xlabel('Size of Training Observations')
plt.ylabel('Accuracy')
plt.legend(loc=0);
```

# 5.8

```python
# 5.8

from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_predict = gnb.predict(X_test)
gnb_score = accuracy_score(y_test, y_predict)
gnb_score
```

```python
# SVC
svc = SVC()
svc.fit(X_train, y_train)
y_predict = svc.predict(X_test)
svc_score = accuracy_score(y_test, y_predict)
svc_score
```

```python
# Decision Tree
dtc = DecisionTreeClassifier(random_state=4444)
dtc.fit(X_train, y_train)
y_predict = dtc.predict(X_test)
dtc_score = accuracy_score(y_test, y_predict)
dtc_score
```

```python
rfc = RandomForestClassifier(random_state=4444)
rfc.fit(X_train, y_train)
y_predict = rfc.predict(X_test)
rfc_score = accuracy_score(y_test, y_predict)
rfc_score
```

# 5.9

```python
# 5.9

from sklearn.cross_validation import cross_val_score

print 'Naive Bayes -', np.mean(cross_val_score(gnb, X, y))
print 'Support Vector Machine -', np.mean(cross_val_score(svc, X, y))
print 'Decision Tree -', np.mean(cross_val_score(dtc, X, y))
print 'Random Forest -', np.mean(cross_val_score(rfc, X, y))
```

The coefficients beta are the log-odds of predicting y=1 (if the coefficient is high) or y=0 (if the coefficient is very negative). Since all the log-odds are close to 0, this means none of the features are particularly good at predicting MPAA rating.
