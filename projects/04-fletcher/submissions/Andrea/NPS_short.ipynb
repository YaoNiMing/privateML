```python
#Now try LDA
from sklearn.feature_extraction.text import CountVectorizer

# Step 1: Create a CountVectorizer for parsing/counting words and fit it using my documents
count_vectorizer = CountVectorizer(ngram_range=(1,2), stop_words='english', token_pattern="\\b[a-z][a-z]+\\b")

count_vectorizer.fit(nps_short['SignificanceComments'])

# Step 2: Create the term-document matrix, and transpose it so the terms are the rows and the columns are how often they appear
counts = count_vectorizer.transform(nps_short['SignificanceComments']).transpose()

counts.shape

#Note to self I would be really tempted to try to improve this by stemming the words...
```

```python
# Next, convert sparse matrix of counts to a gensim corpus
from gensim import corpora, models, similarities, matutils

corpus = matutils.Sparse2Corpus(counts)

id2word = dict((v, k) for k, v in count_vectorizer.vocabulary_.iteritems())
#Note: v is the # and k is the token

len(id2word)
```

```python
# Create lda model (equivalent to "fit" in sklearn) -- try with, say, 20 topics
lda = models.LdaModel(corpus=corpus, num_topics=20, id2word=id2word, passes=10)
```

```python
#Only way I can seem to find to do that is by hand; couldn't figure out how to back out the features for each document from countvectorizer

from nltk.tokenize import word_tokenize, wordpunct_tokenize, RegexpTokenizer
from nltk.util import ngrams
from sklearn.feature_extraction import stop_words
import string
 
#from nltk.corpus import stopwords    
#stops = set(stopwords.words('english'))

tokenizer = RegexpTokenizer("[a-z][a-z]+['\w]+")
stops = set(stop_words.ENGLISH_STOP_WORDS)
additional_stopwords = [",", ".", "'", '"', "?)", "''", ";", ":", "``", "!", "!!", "--", "----", "-", "'s", "an", "is", "(", ")", "?", "its", "isnt", "dont", "doesnt", "he", "his", "my", "ve", "m"]
stops = stops.union(additional_stopwords)

def tokenize_doc(document):
    word_list = []
    doc_no_apost = document.translate(None, "'")
    #word_tokens = word_tokenize(doc_no_punct.lower())
    word_tokens = tokenizer.tokenize(doc_no_apost.lower())
    for word in word_tokens:
        if word not in stops:
            word_list.append(word)
    return word_list

tokens_bypark = nps_short.SignificanceComments.apply(tokenize_doc)
```

```python
nps_short['TokensByPark'] = tokens_bypark
```

```python
#Now prep the corpus
def prep_corpus(docs, no_below=5, no_above=0.5):
  print('Building dictionary...')
  dictionary = Dictionary(docs)
  dictionary.filter_extremes(no_below=no_below, no_above=no_above, keep_n=None)
  dictionary.compactify()

  print('Building corpus...')
  corpus = [dictionary.doc2bow(doc) for doc in docs]

  return dictionary, corpus
```

```python
dictionary, corpus = prep_corpus(nps_short.TokensByPark)
```

```python
#Try Word2Vec with gensim, importing Google's vectors
from gensim import corpora, models, similarities, matutils
import numpy as np
```

```python
model = models.Word2Vec.load_word2vec_format('GoogleNews-vectors-negative300.bin', binary=True)
```

```python
model.save("word2vec_google_news_neg300.model")
```

```python
#Write a function to vectorize nps_short['TokensByPark'], as well as the user input

def get_vectors(word_list):
    feature_vectors = []
    for word in word_list:
        try:
            word_vec = model[word]
            feature_vectors.append(word_vec)
        except:
            pass
        
    feature_averages = []
    for i in range(300):
        vector_to_average = []
        for j in range(len(feature_vectors)):
            vector_to_average.append(feature_vectors[j][i])
        avg_vector = np.mean(vector_to_average)
        feature_averages.append(avg_vector)
        
    return np.array(feature_averages)
```

```python
#NEXT RUN THIS
avg_vector_bypark = nps_short.TokensByPark.apply(get_vectors)
nps_short['avg_vector'] = avg_vector_bypark
```

```python
#Write a function to figure out which park is closest in space to an input string that describes what a 
# hypothetical visitor is looking for

#Note also that Ramesh changed this in the version that is in the flask app, to be a bit more efficient. 
#Look at that code in np-app views.py

from scipy import spatial
from sklearn.metrics.pairwise import cosine_similarity

def park_recommender(string):
    columns=['ParkName', 'cosine_similarity']
    word_list = tokenize_doc(string)
    string_vector = get_vectors(word_list)
    
    cosine_similarity_list = []
    for vector in nps_short.avg_vector:
        cosine_similarity_list.append(cosine_similarity(vector.reshape(1, -1), string_vector.reshape(1, -1)))
    park_vector_pairs = pd.DataFrame(zip(nps_short.ParkName, cosine_similarity_list), columns=columns)
    sorted = park_vector_pairs.sort_values(by='cosine_similarity', ascending=False)
    first, second, third, fourth, fifth = [sorted.iloc[0, 0], sorted.iloc[1, 0], sorted.iloc[2, 0], sorted.iloc[3,0], sorted.iloc[4,0]]
    return first, second, third, fourth, fifth
```
