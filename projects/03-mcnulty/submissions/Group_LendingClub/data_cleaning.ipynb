```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
%matplotlib inline
```

```python
df_accepted = pd.read_csv('LoanStats3d.csv', header=1)
```

```python
df_rejected = pd.read_csv('RejectStatsD.csv', header=1)
```

```python
df_accepted['Accepted'] = 1
df_rejected['Accepted'] = 0
```

```python
df_accepted_merge = df_accepted[['loan_amnt', 'title', 'dti', 'zip_code', 'addr_state', 'emp_length', 'Accepted']]
```

```python
df_accepted_merge.head()
```

```python
df_rejected_merge = df_rejected[['Amount Requested', 'Loan Title', 'Debt-To-Income Ratio', 'Zip Code', 'State', 'Employment Length', 'Accepted']]
```

```python
df_rejected_merge.head()
```

```python
columns = ['loan_amnt', 'title', 'dti', 'zip_code', 'addr_state', 'emp_length', 'Accepted']
```

```python
df_rejected_merge.columns = columns
```

```python
df_rejected_merge = df_rejected_merge.replace('%', '', regex=True)
```

```python
df_rejected_merge.head()
```

```python
df_rejected_merge = df_rejected_merge[~df_rejected_merge.title.isin(['althea9621', 'dougie03', 'freeup', 'loan', 'smmoore2', 'thad31'])]
df_rejected_merge.replace(['debt_consolidation', 'Consolidate debt', 'Consolidation Loan'], 'Debt consolidation', inplace = True, regex=True)
df_rejected_merge.replace(['10 months away from being an RN', 'educational'], 'Educational', inplace = True, regex=True)
df_rejected_merge.replace(['Need a decent rate on car financing', 'Auto Financing', '^car$', 'Car financing'], 'Auto', inplace = True, regex=True)
df_rejected_merge.replace(['Business Advertising Loan', 'Business Line Of Credit', 'Business Loan', 'Small Business Expansion', 'small_business'], 'Business', inplace = True, regex=True)
df_rejected_merge.replace('credit_card', 'Credit card refinancing', inplace = True, regex=True)
df_rejected_merge.replace(['home_improvement', 'house'], 'Home improvement', inplace = True, regex=True)
df_rejected_merge.replace('major_purchase', 'Major purchase', inplace = True, regex=True)
df_rejected_merge.replace('medical', 'Medical expenses', inplace = True, regex=True)
df_rejected_merge.replace('moving', 'Moving and relocation', inplace = True, regex=True)
df_rejected_merge.replace('other', 'Other', inplace = True, regex=True)
df_rejected_merge.replace('renewable_energy', 'Renewable energy', inplace = True, regex=True)
df_rejected_merge.replace('vacation', 'Vacation', inplace = True, regex=True)
```

```python
df_rejected_merge.groupby(by='title').count()
```

```python
df_accepted_merge.replace(['Car financing', 'Credit Card/Auto Repair'], 'Auto', inplace = True, regex = True)
df_accepted_merge.replace(['Pay off Lowes Card', 'Paying off higher interest cards & auto'], 'Credit card refinancing', inplace = True, regex = True)
df_accepted_merge.replace('New Baby and New House \(CC Consolidate\)', 'Moving and relocation', inplace = True, regex = True)
df_accepted_merge.replace('Prescription Drug and Medical Costs', 'Medical expenses', inplace = True, regex = True)
df_accepted_merge.replace('DebtC', 'Debt consolidation', inplace = True, regex = True)
df_accepted_merge.replace(['Learning and training', 'Student Loan'], 'Educational', inplace = True, regex = True)
df_accepted_merge.replace(['SAVE', 'Simple Loan Until Contract Is Completed', 'Trying to come back to reality!', 'considerate', 'new day', 'new kitchen for momma!', 'odymeds'], 'Other', inplace = True, regex = True)
```

```python
df_accepted_merge.groupby(by='title').count()
```

```python
df_total = pd.concat((df_accepted_merge, df_rejected_merge), ignore_index = True)
```

```python
df_total.dti = df_total.dti.astype(float)
```

```python
df_accepted_merge.groupby(by='emp_length').count()
```

```python
df_total = df_total.dropna()
```

```python
df_total.to_csv("loan_app_data.csv", index=False)
```
