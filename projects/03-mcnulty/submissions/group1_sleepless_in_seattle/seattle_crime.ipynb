```python
%%time
df.loc[:, 'Event Clearance Date'] = df['Event Clearance Date'].apply(lambda x:dateutil.parser.parse(x))

df.loc[:, 'At Scene Time'] = df['At Scene Time'].apply(lambda x:dateutil.parser.parse(x))
```

```python
df['ClearanceTime'] = df['Event Clearance Date'] - df['At Scene Time']
```

```python
df['ClearanceTime'] = df['ClearanceTime'].apply(lambda x:x.seconds)
```
