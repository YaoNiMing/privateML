```python
#Request data from IMDB and save the files in folder 'business'

import time

for i in omdb_df['imdbID']:
    url = "http://www.imdb.com/title/" + i + "/business?ref_=tt_dt_bus"
    response = requests.get(url)
    if response.status_code == 200:        
        filename = "business/" + i + ".html"
        #Now save that file
        with open(filename, 'w') as f:
            f.write(response.text.encode('utf-8'))
    else:
        continue      
    time.sleep(.3)        
    
    
```

```python
#Search through all the html files and return a dictionary that contains all the available financial info for each
import os  
import re

id_dict = {}

for file in os.listdir('business/'):
    soup = BeautifulSoup(open(os.path.join('./business/',file)).read(), 'lxml')
    #file = "tt0081383.html"
    div = soup.find(lambda tag: tag.name=='div' and tag.has_attr('id') and tag['id']=="tn15title")
    id = div.find('a')['href'].split('/')[2]

    div2 = soup.find(lambda tag: tag.name=='div' and tag.has_attr('id') and tag['id']=="tn15content")
    div2 = str(div2)   
    info = re.findall('<h5>([\w\s]+)<\/h5>\n(.*?)<br\/>', div2)
    
    #<h5>(\w+)<\/h5>\n(.*?)<br\/>

    id_dict[id] = info
```

```python
#Now search through this dict for useful info and create a dataframe

import re
movie_amounts = []
for movie_id, values in id_dict.items():
    movie = {}
    movie["imdbID"] = movie_id
    for amt_key, amt_value in values:
        if amt_key == "Budget" or amt_key == "Gross" or amt_key == "Opening Weekend":
            matched_val = re.match(r"(.*?[\d,]+)", amt_value)
            if matched_val:
                movie[amt_key] = matched_val.group(1)
            matched_usd = re.match(r".*?(\$[\d,]+)", amt_value)  
            if matched_usd:
                movie[amt_key + "_usd"] = matched_usd.group(1)
    movie_amounts.append(movie)
```

```python
#3. Deal with parsing the awards column.  Here create a function to sum awards, nominations, and awards + noms.
# Returns all 3 in their own lists

import re

def awards(column):
    noms = []
    awards = []
    noms_awards = []
    for cell in column:
        cell = str(cell)
        
        nominatedfor = re.search(r'for (\d*)', cell)
        more_nominations = re.search(r'(\d*) nomination', cell)
        
        if nominatedfor and more_nominations:
            all_nominations = int(nominatedfor.group(1)) + int(more_nominations.group(1))
        elif nominatedfor and not more_nominations:
            all_nominations = int(nominatedfor.group(1))
        elif more_nominations and not nominatedfor:
            all_nominations = int(more_nominations.group(1))
        else:
            all_nominations = 0
            
        won = re.search(r'Won (\d*)', cell)
        wins = re.search(r'(\d*) win', cell)
        
        if won and wins:
            all_awards = int(won.group(1)) + int(wins.group(1))
        elif won and not wins:
            all_awards = int(won.group(1))
        elif wins and not won:
            all_awards = int(wins.group(1))
        else:
            all_awards = 0    
        
        nom_plus_award = all_nominations + all_awards
        
        noms.append(all_nominations)
        awards.append(all_awards)
        noms_awards.append(nom_plus_award)
  
    return noms, awards, noms_awards        
            
noms, awards, noms_awards = awards(final_df['Awards'])            
```
