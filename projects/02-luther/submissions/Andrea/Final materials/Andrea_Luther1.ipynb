```python
#Create a function that returns a list of dictionaries to hold information on each original and remake

def get_pairs(file):
    movie_pairs = []
    header = ['remake', 'original']
    for row in file.findAll('tr'):
        row_dict = {}
        for i, cell in enumerate(row.findAll('td')):
            row_dict[header[i]] = cell.findAll(text = True)
        movie_pairs.append(row_dict)

    # Eliminate empty dictionaries (from the headers of the tables by first letter of movie in Wikipedia)    
    more_pairs = []
    for i in movie_pairs:
        if i:
            more_pairs.append(i)
    movie_pairs = more_pairs
    
    return movie_pairs   
```

```python
#Call get_pairs on my 2 html files from wikipedia
atom = BeautifulSoup(open('movies_atom.html').read(), 'lxml')
pairs_atom = get_pairs(atom)
ntoz = BeautifulSoup(open('movies_ntoz.html').read(), 'lxml')
pairs_ntoz = get_pairs(ntoz)
```
