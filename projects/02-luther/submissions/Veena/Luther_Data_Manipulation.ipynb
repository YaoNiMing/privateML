```python
def convert_to_num(dataframe,column):
    for i in range(len(dataframe[column])):
        row = dataframe[column][i]
        if row != 'None':
            num = row.replace('$','').replace(',','')
            num = int(num)
            return num
        else:
            return 0

for i in range(len(movie_data)):
    row = movie_data['Release Yr'][i]
    rate = current/adjusted[int(row)]
    movie_data['Domestic Total'][i] = int(movie_data['Domestic Total'][i])*rate
return movie_data['Domestic Total'][i]
```

```python
#Converting currency columns into int.

for i in range(len(movie_data['Domestic Total'])):
    row = movie_data['Domestic Total'][i]
    if row != 'None':
        row = row.replace('$','').replace(',','')
        movie_data['Domestic Total'][i] = int(row)        
    else:
        movie_data['Domestic Total'][i] = 0

for i in range(len(movie_data['Foreign:'])):
    row = movie_data['Foreign:'][i]
    row = str(row)
    if row != 'None' and row !='nan' and row !='n/a':
        row = row.replace('$','').replace(',','')
        movie_data['Foreign:'][i] = int(row)        
    else:
        movie_data['Foreign:'][i] = 0
```

```python
#Extracting only release year. 
movie_data['Release Yr'] = [row[-4:] for row in movie_data['Release Date']]
```

```python
#Changing type from string to int.
movie_data['Release_Yr'] = map(int,movie_data['Release_Yr'])
```

```python
#Adjusting for inflation.
current = 8.66
adjusted = {2000: 5.39, 2001: 5.66, 2002: 5.81, 2003: 6.03, 2004: 6.21, 2005: 6.41, 2006: 6.55, 2007: 6.88, 2008: 7.18, 2009: 7.50, 2010: 7.89, 2011: 7.93, 2012: 7.96, 2013: 8.13, 2014: 8.17, 2015: 8.43, 2016: 8.66}

years = pd.unique(movie_data['Release_Yr'])

for yr in years:
    criteria = movie_data['Release_Yr'] == yr
    movie_data.Foreign[criteria] = movie_data.Foreign[criteria]*current/adjusted[yr]
```

```python
#Converting Production Budget into a number.

#Step 1 : get rid of the 'million'
for i in range(len(movie_data['Production_Budget'])):
    movie_data['Production_Budget'][i] = str(movie_data['Production_Budget'][i])
    if movie_data['Production_Budget'][i].endswith('million'):
        movie_data['Production_Budget'][i] = movie_data['Production_Budget'][i].split()[0]

#Step 2 : strip and replace '$' 
for i in range(len(movie_data['Production_Budget'])):
    row = str(movie_data['Production_Budget'][i])
    if row != 'None' and row != 'nan':
        row = row.replace('$','').replace(',','')
        movie_data['Production_Budget'][i] = int(float(row)) * 1000000
    else:
        movie_data['Production_Budget'][i] = 0
#Step 3 : Multiply value by 1 million to get the correct value.

```

```python
import os
path = '/Users/veenakumar/Desktop/Projects/Luther/BomData/'

def runtime_to_minutes(runtimestring):
    runtime = str(runtimestring)
    runtime = runtime.split()
    try:
        minutes = int(runtime[0])*60 + int(runtime[2])
        return minutes
    except:
        return None
          
def runtime(df):
    for i in range(len(df['Runtime'])):
        df['Runtime'][i] = runtime_to_minutes(df['Runtime'][i])
```

```python
def build_genre_indicators(row):
    try:
        movie_genres = row["Genre"].lower().split(',')
        movie_genres = [g.strip() for g in movie_genres]
        for genre in ["comedy", "romance", "action", "drama", "family","thriller","short","sci-fi"]:
            if genre in movie_genres:
                row[genre] = 1
        return row
    except:
        return None
```
