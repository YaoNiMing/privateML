```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import urllib
import re
import requests
import json

from bs4 import BeautifulSoup

%matplotlib inline
```

```python
from os import listdir
from os.path import isfile, join
mypath = "BoxMojoHTML/"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
onlyfiles.pop(0)
print onlyfiles
```

```python
movies = []
for movie in range(0, len(onlyfiles)):    
    url = "BoxMojoHTML/" + onlyfiles[movie]
    page = urllib.urlopen(url).read()

    soup = BeautifulSoup(page)
    title = soup.find_all('b')[1]
    title = title.text
    title = title.split(" ")
    link = "http://www.omdbapi.com/?t="
    for words in range(0, len(title)):
        link = link + str(title[words])
        if (words + 1) <= len(title):
            link = link + "+"
    link = link + "&tomatoes=true&r=json"

    movieresponse = requests.get(link)
    
    moviestring = movieresponse.text.encode('utf-8')  
    if '"Response":"False"' not in moviestring:
        movies.append(movieresponse.json())
pd.DataFrame(movies).to_csv('MoviesIMDB.csv', encoding='utf-8')
```
