```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import urllib
import re
import requests

from bs4 import BeautifulSoup

%matplotlib inline
```

```python
from os import listdir
from os.path import isfile, join
mypath = "BoxMojoHTML/"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
onlyfiles.pop(0)
```

```python
df = pd.read_csv('MoviesIMDB.csv')
df.drop(['Unnamed: 0','Awards','BoxOffice','DVD','Language','Metascore','Plot','Poster','Response','Runtime','Type','Website'], axis = 1)
```

```python
movielist = []
moviegross = []
for movie in range(0, len(onlyfiles)):
    url = "BoxMojoHTML/" + onlyfiles[movie]
    page = urllib.urlopen(url).read()
    soup = BeautifulSoup(page)
    title = soup.find_all('b')[1]
    title = title.text
    domestic_total_regex = soup.find(text=re.compile('Domestic Total Adj. Gross: '))
    domesticgross = domestic_total_regex.findNextSibling()
    movielist.append(title)
    moviegross.append(domesticgross.text)
```

```python
columns = ['Title', 'Gross Adj Profit']
df2 = pd.DataFrame(movielist, moviegross).reset_index()
df2.columns = ['Gross Adj Profit', 'Title']
```

```python
df2.head()
```

```python
moviesdf = pd.merge(df, df2, left_on='Title', right_on='Title')
moviesdf.to_csv("MOVIESDF.csv")
```
