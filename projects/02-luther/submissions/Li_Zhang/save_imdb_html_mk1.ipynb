```python
import os
import re
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests
import re
import time
import numpy as np
```

```python
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_8_5)'}

imdb_local = 'IMDb_ Most Popular Feature Films Released No Later Than 2015-12-31 - IMDb.htm'

with open(imdb_local, 'r') as f:
        file_content = f.read()
```

```python
bs = BeautifulSoup(file_content, 'lxml')
```

```python
re_title = re.findall('<a href="/title.*/\?ref', file_content)
```

```python
len(re_title)
```

```python
title = re_title[0::2]
```

```python
len(title)
```

```python
# find all ids
web_id = []
for title_1 in title:
    web_id.append(re.findall('tt\d+', title_1)[0])
```

```python
# save all html files

headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_8_5)'}

counter = 0

for mov_id in web_id[4190:]:

    counter += 1
    
    # sleep
    time.sleep(np.random.random()*0.25)
    print(str(counter) + '   saving ' + mov_id + '...')

    # retrieve web
    url_mov = 'http://www.imdb.com/title/' + mov_id + '/?ref_=adv_li_tt'
    response_mov = requests.get(url_mov)
    response_text = response_mov.text

    # write local
    html_name = mov_id + '.html' 
    with open('/Users/NoSlack/sf16_ds4/local/Luther/imdb/' + html_name, 'w') as f:
        f.write(response_text)

```
